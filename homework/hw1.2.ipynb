```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.2)
y = x**3 - x**2 + (2*x) + 5
noise = 100 * np.random.normal(size=x.size)
ydata = y + noise
plt.plot(x, ydata, 'ro')
plt.plot(x, y, 'g')
plt.show()
```

```python
# using the least square regression method

A = np.vstack([x, np.ones(len(x))]).T
by = ydata[:, np.newaxis]
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), ydata)
print(alpha)
plt.plot(x, ydata, 'r.')
plt.plot(x, alpha[0]*x + alpha[1], 'g.')
plt.show()
```
